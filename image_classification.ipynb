#image classifiaction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
img=imread('/content/drive/MyDrive/data/Cat/cat.11.jpg')  #to read he image
img
```

```python
data_array=[]  #to store image
target_array=[]  #to store the classifiaction
catergories=['Cat','Dog']

datadir='/content/drive/MyDrive/data'


for i in catergories:
  sub_path=os.path.join(datadir,i)
  images=os.listdir(sub_path)
  for img in images:
     img_path=os.path.join(sub_path,img)
     img_array=imread(img_path)
     img_resized=resize(img_array,(150,150,3))   #we gave a shape to resize the data were we have 150 lements with each 150*3 matrix
     data_array.append(img_resized.flatten())    #flatten means all the elements in single
     target_array.append(catergories.index(i))
  print('Loaded catergory:',i)
```

```python
data=np.array(data_array)
data.shape   #150*150*3
```

```python
a=np.zeros([2,2,3]) #2*2*3 =12
a.shape
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df  #all the datas are scaled when it is resized  0-255(RGB) each value divided by 255
```

```python
df['target']=target
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[::,-1]
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)

```

```python
svm=SVC(kernel='linear')
svm.fit(X_train,y_train)
y_pred=svm.predict(X_test)
y_pred
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
#The accuracy of the model is 85% which is low.it is because the number of images used for training the model was just 126.For better performance training the model
# with more number of data is good for more accuracy od the data set.
```
